## Основы работы с Pandas. Связывание таблиц

Данные часто поставляются в разных источниках. Чтоб их сопоставит, необходимо их объединить

### Объединение таблиц     
#### .merge()

```python
import pandas as pd  #Подключаемся к библиотеке
```

```python
mov=pd.read_excel('movement.xlsx') # Считываем
```

```python
mov.head(5) # Проверяем, что считали
```

```python
med=pd.read_excel('medication.xlsx')
```

```python
med
```

```python
phar=pd.read_excel('pharmacies.xlsx')
```

```python
phar
```

> Метод .merge() позволяет объединять данные по общему ключу

### Типы соединений в Pandas

1. **`inner`** — работает по умолчанию и выводит значения, ключ по которым встречается и в первой, и во второй таблице.
2. **`left`** — выводит все значения из первой таблицы и значения из второй таблицы, к которым нашлись ключи из первой.
3. **`right`** — выводит все значения из второй таблицы и значения из второй таблицы, к которым нашлись ключи из второй (лучше его не использовать).
4. **`outer`** — выводит все значения из обеих таблиц и совпадающие ключи и уникальные для первой или второй таблицы.

! Для всех отсутствующих значений присваивается NaN.

```python
df_new=pd.merge(mov, phar, 
                how='left', 
                left_on='ID аптеки', right_on='ID аптеки')
```

```python
df_new
```

```python
df_full_merge=pd.merge(df_new, med, 
                how='left', 
                left_on='Артикул', right_on='Артикул')
```

```python
df_full_merge
```

```python
pilylia_filtrum = df_full_merge[(df_full_merge['Сеть'] == 'Пилюля') & (df_full_merge['Название'] == 'фильтрум')]
pilylia_filtrum
```

```python
pilylia_filtrum['Цена итого'] = pilylia_filtrum['Количество упаковок, шт.'] * pilylia_filtrum['Цена за упаковку, руб.']
```

```python
pilylia_filtrum
```

```python
result = pilylia_filtrum['Цена итого'].sum()
```

```python
result
```
