В файле приведён фрагмент базы данных **«Кондитерские изделия»** о поставках конфет и печенья в магазины районов города. База данных состоит из трёх таблиц.

Таблица «Движение товаров» содержит записи о поставках товаров в магазины в течение первой половины августа 2023 г., а также информацию о проданных товарах. Поле Тип операции содержит значение Поступление или Продажа, а в соответствующее поле Количество упаковок, шт внесена информация о том, сколько упаковок товара поступило в магазин или было продано в течение дня.

Используя информацию из приведённой базы данных, определите **магазин на Мартеновской улице** с наибольшей выручкой от продажи **пряников всех видов за вторую половину августа**. В ответе укажите одно число — найденную выручку от реализации пряников для этого магазина.

Т.к. я знаю, что данных не так много, то я лучше сначала соединю несколько таблиц, а на их основе уже произведу фильтрацию.

```python
import pandas as pd

move = pd.read_excel('move.xlsx')
move
```

```python
product = pd.read_excel('product.xlsx')
product
```

```python
shop = pd.read_excel('shop.xlsx')
shop
```

Объединим move и product и shop

```python
move_product = pd.merge(move, product, on='Артикул', how='left')
move_product_shop = pd.merge(move_product, shop, on='ID магазина', how='left')
move_product_shop
```

Теперь можем отфильтровать датафрейм по нужным нам данным

```python
move_martenovskaya = move_product_shop[move_product_shop['Адрес'].str.contains('мартеновская', case=False, na=False)]
move_martenovskaya
```

Теперь отфильтруем операции по пряникам

```python
move_martenovskaya_prianyki = move_martenovskaya[move_martenovskaya['Наименование товара'].str.contains('пряник', case=False, na=False)]
move_martenovskaya_prianyki
```

```python
move_martenovskaya_prianyki_august = move_martenovskaya_prianyki[move_martenovskaya_prianyki['Дата'].between('2023-08-16', '2023-08-31')]
move_martenovskaya_prianyki_august['Итого'] = move_martenovskaya_prianyki_august['Количество упаковок, шт'] * move_martenovskaya_prianyki_august['Цена за упаковку']
move_martenovskaya_prianyki_august
```

```python
group_byShops = move_martenovskaya_prianyki_august.groupby('Адрес').sum(['Итого'])
group_byShops
```

```python
result = group_byShops.max()
result
```
