```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
self_d = {
    'Муниципальное_образование' : ['Братск', 'Усть Илимск', 'Бодайбо', 'Усть Кут'],
    '2020' : [2641.7, 845, 808, 359],
    '2021' : [3190, 1011, 908.4, 369],
    '2022' : [3328.4, 1089, 940, 675],
    '2023' : [3535.8, 1203, 974, 654]
}

not_self_d = {
    'Муниципальное_образование' : ['Братск', 'Усть Илимск', 'Бодайбо', 'Усть Кут'],
    '2020' : [5231.9, 2066, 417.8, 308],
    '2021' : [6579.3, 2400, 408, 458],
    '2022' : [7703.1, 2971, 412, 1558],
    '2023' : [9513.4, 3310, 366, 1487]
}
```

```python
df_self = pd.DataFrame(self_d)
df_not_self = pd.DataFrame(not_self_d)
df_self
df_not_self
```

```python
years = ['2020', '2021', '2022', '2023']

municipalities = self_d['Муниципальное_образование']

x = np.arange(len(years))

bar_width = 0.2

fig, ax = plt.subplots(figsize=(10, 10))
colors = ['blue', 'green', 'orange', 'purple']

for i, municipality in enumerate(municipalities):
    offset = (i - 1.5) * bar_width
    # Собственные доходы
    self_values = [self_d[year][i] for year in years]
    self_bars = ax.bar(x + offset, self_values, 
                       width=bar_width, label=f'{municipality} Собственные доходы, млн. руб.', color=colors[i], alpha=0.5)
    
    # Безвозмездные поступления
    not_self_values = [not_self_d[year][i] for year in years]
    not_self_bars = ax.bar(x + offset, not_self_values, 
                           width=bar_width, label=f'{municipality} Безвозмездные поступления, млн. руб.', color=colors[i], alpha=0.7)
    
    # Добавление значений над столбцами
    for self_bar, not_self_bar, self_value, not_self_value in zip(self_bars, not_self_bars, self_values, not_self_values):
        # Значение для self_d
        ax.text(self_bar.get_x() + self_bar.get_width() / 2, 
                self_bar.get_height() + (20 if self_bar.get_height() <= not_self_bar.get_height() else 80), 
                f'{self_value:.1f}', ha='center', va='bottom', fontsize=8)
        # Значение для not_self_d
        ax.text(not_self_bar.get_x() + not_self_bar.get_width() / 2, 
                not_self_bar.get_height() + (80 if not_self_bar.get_height() > self_bar.get_height() else 20), 
                f'{not_self_value:.1f}', ha='center', va='bottom', fontsize=8)


ax.set_xlabel('Годы')
ax.set_ylabel('Значения')
ax.set_title('Сравнение показателей муниципальных образований')
ax.set_xticks(x)
ax.set_xticklabels(years)

# Настройка положения легенды
ax.legend(loc='lower center', bbox_to_anchor=(0.5, -0.2), ncol=2)

# Уменьшение плотности элементов графика
plt.tight_layout()
plt.grid()

plt.show()
```
